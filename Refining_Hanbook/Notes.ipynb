Every refinery has it's own refining capacity either expressed in ***BPSD*** or ***BPCD***
- ***BPSD*** : "Barrels Per a Stream Day" this means  how many barrels at a day fully operational (No down time)
- ***BPSCD*** : "Barrels Per Calendar Day" this means how many barrels per normal streaming capacity (no down time)
BPSD > BPCD

## Classification of refinning operations
There are 4 types : 
- **Distillation** : Atmospheric and vaccum columns to separate the crude into groups of hydrocarbon compounds based on size and boiling point range.
- **Conversion Processes**: These are to change the size and structure of hydrocarbon molecules by : 
 1. Decmposition : Through cracking and similar processes
 2. Unification : Alkylation, Polymerisation
 3. Reforming : Isomerisation, Catalytic reforming

- **Treatment processes**: Prepare Hydrocarbon strams for additional processes and to prepare finished products using chemical and physical separations "Desalting, hydrodesulfurization, solvant refinning, sweetening, solvant extraction and dewaxing"
- **Blending**: The process of mixing and combining Hydrocarbon fractions, additives to produce finished products.

## Important charachterization properties
*SpGr*, *%S*, *TAN* these are the most important properties in charachterizing crude oil.
- **SpGr*: Specific gravity is the density of the crude and is measure in °API. The higher the °API the lower the desnity ($\frac{Kg}{\text{m}^3}$) $\rightarrow$ paraffinic $\rightarrow$ and give high proportion yields of gasoline and light petroleum products.

- *%S*: This represents the sulfure percentage. Sulfure is an undesirable impurity because of pollution and corrosion concerns. it is measure in $\text{wt\%}$ 
$$\uparrow \text{s\%} \rightarrow sour$$ and also $$\downarrow \text{s\%} \rightarrow sweet$$
and the crossover is at (1.5% - 0.5%)

- *TAN*: Total Acid number, this measure the potential of corrosivity. It quantifies the milligrams of potassium hydroxxide needed to neutalize 1g of crude. therefore it is expressed in : $$TAN: \frac{mg(KOH)}{g(Crude)}$$
Corrosive if TAN > 1 $\frac{mg(KOH)}{g(Crude)}$

## Types of refineries:

- Simple : Atmospheric distillation, catalytic reforming and one or more pretreatment facilities and hydrodesulfurization unit.
- Complex : Vaccum distillation, catalytic reforming and catalytic cracking.
- More Complex : Includes coking processes

The refinning process is an energy intensive process, where 35-40% of the energy is used in by the atmospheric and vaccum distillation. and as most refinery streams, hydrotreating is necessary which takes around 20%.

### Properties of crude oil
#### API gravity
A measure of density and related to SpGr : $$\text{°API} = \frac{141.5}{SpGr} - 131.5$$
$10 \leq °API \leq 50$  (Could be over 50)

#### Flash point 
The temperature at which the vapor above the oil will momentarily flash or explopde

#### Octane number 
This is a measure of a gosoline's engine resistance to knock or detonation in a cylinder of the engine.
By definition, the octane number is a percentage of iso-octane in a blend of iso-octane and n-heptane.

#### Viscosity
Is the measure of an oil resistance to it's internal flow, and is an indication to the oil lubricating qualities.

#### Could and pour points
The are tests that indicate the relative coagelation of wax in the oil (indication of the paraffinity or aromaticity of the oil)
$$\downarrow \text{pour point} \downarrow \text{paraffinic content}  \uparrow \text{aromatic content} $$

#### Sulfur content
Quantity of sulfur in oil, sulfur can be present in oil as: 
- Mercaptans : R-SH
- Disulfides : R-S-R
- Thiophenes : C4H4S

#### Carbon Residu
Carbon residu is determined by distillation to coke in the absence of air. It's roughly related to the asphalt content and the lubricating oil fraction that can be recovered. The lower the carbon residu teh more valuable the crude.

#### Salt content
If the salt content of crude is greater than 10lb/1000bbl. Generally it is necessary to desalt the crude before treatment, otherwise severe corrosion problems maybe encountered
$$ \frac{10lb}{1000bbl} \rightarrow 3PPM $$

#### Characterization factors

$ \text{Kw} = \frac{Tb^\frac{1}{3}}{SpGr^60} $ ; $ \text{Kw} = \frac{87.552}{Tb} +473.7G - 456.8 $

- Kw (Kuop) rand from less than 10 for highly aromatic materials to almost 15 for highly paraffinic materials.
- CI index is useful for evaluating individual fractions of crude oils. It's based on straight chain paraffines having CI value of 0 and benzene having CI index value of 100

#### TBP curve
It is the plot of the boiling points of almost pure components or fractions contained in the crude oil

#### ASTM distillation curve
It is used on routine basis for plant and product quality control

#### Calculating the SpGr of a cut
You get the weight factor by multiplying the SG by the % of volume. Then you devide the weight factor by the sum of all the fractions.
$$ SG = \frac{\sum{fi*SGi}}{\sum{fi}} $$

#### Calculating the sulfur content
You determine the sulfur factor (sulfur% * weight factor) the you devide by the sum of weight fractions
$$ \%S = \frac{\sum{fi*SGi*Si}}{\sum{fi*SGi}} * 100 $$

### Crude desalting 

If salt content is greater than 3 PPM (NaCl) $\rightarrow $ desalting is necessary to minimize fouling and corrosion caused by salt deposition on heat transfer surfaces and acides formed by decomposition of chloride salts.
Some metals in inorganic compounds dissolved with water emulsified with crude oil which can lead to catalyst deactivation are also rejected in the desalting process.

The salt in the crude is in the form of dissolved or suspended salt crystals in water emulsified with the crude oil which leads to problems that occur in : 
- Obtaining efficient and economical water/oil mixing.
- Water wetting of suspended solids.
- Separation of the wash water from the oil

The separation ease and efficiency are affected by : ***Ph, gravity, viscosity and the volume % of wash water used per volume of crude***.

***Some aspects***: 
- Efficient and economic water/oil mix (3-10% water at 90-150°C)
- Wetting agents (oxyalkylated, Alkylphenols)
- Separation (Chemical agents or current)

### Atmospheric topping unit
After desalting, the crude oil is pumped through a series of heat exchangers and it's temperature is raised to 288°C by heat exchange with products and reflux streams. Then it's further heated to 399°C in a frunace and charged to the flash zone of the atmospheric distillation tower.
The furnace discharge temperature is sufficiently high to cause vaporization of all products withdrawn above the flash zone

### Vaccum distillation
The temperature required for the heavier fractions at atmospheric pressures are so high that thermal cracking would occur before they are distilled. That's why we use a vaccum because with a drop in pressure the boiling temperature also drop.

The addition of steam improve voprization to the furnace inlet at the bottom of the vaccum tower (also minimizing coke formation)

The effective pressure at the flash zone determine the fraction of feed vaporized at a given outlet temperature. That's why the ***Fractionation tower, overheadlines and condesor*** are designed to minimize the pressure drop between the vaccum inducing device and the flash zone. 
The **Pressure drop** can significantly increase in the volume of the vapor per barrel vaporized, that's why vaccum distillation colymnns are larger in diameter than atmospheric ones (can reach up to 12.5m)

### Catalytic reforming
Catalytic reforming is a process where light petroleum distillates (naphta) are contacted with a platinum containing catalyst at elevated temperature and elevated hydrigen pressure for the purpose of raising their octane number.
The reformate stream can either be used as high octane gasoline blending component or as a source of aromatics. 

***Reforming reactions***:
- Aromatization through dehydrogenation of naphtenes
- Cyclization and dehydrogenation of paraffines to the analogue aromatics.

***Octane numbers*** of hydrocarbons as hydrocarbon mixtures are determined by comparing their anti-Knock properties to various blends of n-heptane(0), 2,2-4 trimethyl pentane (Iso-octane, 100)
Hydrocarbons may apprear to have different octane when blended with other mixtures, this is called *Blending Octane*.
In the reforming reactions, the catalyst used (Platinum and alumina) are bifunctional whereas separate and distinct reactions occur on the platinum side than the alumina.

The platinum side typically performs dehydrogenation and hydrogenation whereas on the acidic alumina side cyclization, isomerization and cracking are performed.
Paraffin isomerization is the most difficult step in reforming, so the ability to convert paraffin selectivey is of paramount importance.

Another acidic catalyized reaction isthe cracking of paraffins into light products. Thus improving octane number by turning them into light gaseous low molecular weight paraffins.
Paraffins can also be cyclized into naphtenes, this reaction can be realized through an olefinic intermediate. 

### Summary of reactions
***Hydrogenation, dehydrocyclyzation, isomerization and hydrocracking*** : 
- Saturated olefines with already paraffines are isomerized and to some extent converted to naphtenes and then subsequently converted into aromatics.
- Naphtenes directly converted to aromatics through dehydrogenation.

***Note*** : The reforming reaction is endothermic so we must introduce heat to the mixture before entering each reactor

***Reactions leading to undesirable products*** : 
- Dealkylation of side chains on naphtenes and aromatics forming butane and light end paraffins
- Cracking of paraffines and naphtenes to form butane and lighter paraffines

## Catalytic Cracking
The purpose of this process is to transform heavy oils into more valuable gasoline and lighter products. The gases produced by this type of cracking contain more olefins than thermal cracking. 
The process produces coke which leads to deactivation of catalyst. So to regenerate the catalyst we burn off the catalyst with air. So the catalyst is moved back and forth from the reactor into the regenarator and vice versa.

The cracking reaction is endothermic and the regeneration reaction is exothermic so it is possible to establish a heat balance to conserve the energy. "The generator temperature is controlled to avoid overheating and therefor the total decay of the catalyst and also to provide the desired amount of carbon burn-off".

### The fluid catalytic cracking 
The FCC process employs the catalyst in the form of very fine particles, which behave as a fluid when aerated with vapor. There are two types of reactors, The ***Side by Side*** and ***OrthoFlow***.

- The fresh feed and the recycled streams are preheated by heat exchanges or furnaces to enter the unit at the base of the riser. Where they are mixed with the hot catalyst, this heat vaporizes the streams of feed and brings it up to desired reactor temperature. 
This mixture trabels up the riser to the separator, the cracking reaction starts when the feed meets with the hot catalyst in the riser and continue until the catalyst and hydrocarbon vapor are separated in the reactor-separator section.

The product is sent to the fractionation tower for seperation into liquid and gaseous products. The spent catalyst is sent to the regenerator, which it contains hydrocarbons as well as coke adsorbed on it's internal surfaces as well as external ones ***(Regenerator T ~ 620 - 845°C)***.

- Zeolite cristalline alumina silica catalyst have a much higher cracking activity than amorphus catalyst and shorter reaction time are required to prevent overcracking of gasoline into gas and coke.
The catalyst activity loss due to metals is mainly caused by vanadium, where increased coke and hydrogen are due to nickel deposited on the catalyst.
For feed containing about 15PPM NiV, passivation of nickel catalyst can be acheived by the addition of organic antimony or bismuth compounds to the feed (Reduction of coke and hydrogen yeild by 15-45%)

For feed containing > 15PPM metals and having **Conradson Carbon** Content > 3%, hydroprocessing may be necessaty of the feed to remove metals and reduce carbon forming potential and increase the yeild of gasoline and middle distillates

***Note*** : 
- Conrandso carbon (Or conradson carbon residue CCR) is a measurement of the tendency of petroleum product to form coke. It quantifies the amount of carbonaceous residue left after a sample is heated under specific conditions, indicating it's potential to leave deposits.

### Cracking reaction
The prodcuts of catalytic cracking are both result of primary or secondary reactions.
The primary reactions are : 
- Paraffin $\rightarrow$ paraffin + Olefine
- AlkylNaphtene $\rightarrow$ Naphtene + Olefine
- Alkyl Aromatic $\rightarrow$ Aromatic + Olefin

## Thermal cracking units 
These processes are cheap relatively compared to FCC or hydrocracking. But they go a long way of acheiving the heavy oil cracking objective of converting low quality feed into valuable oil products.
There are three thermal cracking operations : 
- Thermal crackers (Atmospheric residue)
- Visbreaking (Vacuum residue)
- Cokers (Long or short residues)

The term thermal cracking is given to those processes that convert heavy oils into lighter feed stock such as LPG, Naphta and middle distillates. By applying only heat to the feed over a prescribed period of time.
These processes have the same configuration of processes such as : 
- A cracking furnace
- Soaking vessel or coil
- Product recovery fractionator(s)

The feed is first preheated, then the cracking furnace brings it to it's determined cracking temperature for a predetermined period of time. The cracked oil is quenched by a cold heavy oil stream on leaving the soaking section to a temperature below it's cracking temperature. It then enters the fractionation towers for the products to be separated.

## Gas treating processes
The gas treating processes are essential to remove **$H_2S$** impurities generated by hydrotreating or cracking from the refinery fuel gas or hydrogen recycle stream. THe removal is accomplished by absorbing the 
$H_2S$ and $H_2$ into an amine or similar solution that's ready to absorb $H_2S$. Then stripping the absorbed from $H_2S$ and reacting the latter with air to produce elemental sulfur. 
The rich $H_2S$ gases from all refinery sources enter below the bottom tray of the absorber tower. The lean $H_2S$ absorbent solution enters the tower above the top tray to move down the tower counter current to the gas moving upward. 

The $H_2S$ gas free leaves from the top of the tower and routed to be used as refinery fuel or other prescribed destination. 
The rich absorbent solution leaves the bottom to be heated through heat exchange before entering the stripping columns. The vapor (rich in $H_2S$) goes straight into the sulfur plant.

Petroleum refineries work to meet with "Clean Air" legislations by reducing the pollutants.

Oxygenates are ether compounds derived from their respective alcohols. 
There are three candidates :

- Ethyl tertiary butyl ether (ETBE)
- Tertiary amyl methyl ether (TAME)

### Production of ethers

The olefins feed from FCCU on a steam cracker is combined with methanol stream to enter a guard reactor to remove impurities, a small hydrogen stream is added to the hydrocarbon from the guard reactor prior to entering the ether reactor. 

This Reactor contains special resin catalyst and the reactor feed flows upward through this catalyst bed at moderate temperatures and pressures and in the liquid phase. The reaction is exothermic so temperature is controlled externally by recycle stream from the first two reactors. 
The catalyst in this case performs three reactions simultaneously : 
- Etherfication of branched olefines
- Selective hydrogenation of unreacted di-olefins
- Hydro-isomerisation of olefin by a double bond switch

## Non energy refineries
### Lube oil refinery 
Only about 8 or 9 base lube oil stocks are produced from refinery streams. The many hundresds of commercial grades of lubricating oils are blends of these stocks with small portions of proprietary additives (Mostly acid derivitives)

Two important by products to lube oil products are ***Bitumen and Waxes***.

Lube oil production starts with vacuum distillation to the atmospheric residue. This feedstock is cut to three distillate streams. 
- The lighter stream is called spindle and distilled under vacumm to three blending cuts. These are called spindle oils and form the basis of blending stocks for lubes used domestically or light motor oil. They require little treatment, usually some hydrotreating suffices color specification. 
- The second distillate is dewaxed and sent to engine lube oil pool, it also may be blended with heavier side streams as heavy oil engine stock. 
- The bottom side streams is one of base blending stocks for heavy engine oils and turbine oil stocks. 

The removal of undesired products could be done by solvent extraction, which is accomplished prior to the stream being dewaxed.

The heavy vacuum residue is routed to propane de-asphalting unit (removal of heavy bitumen asphaltene by extraction with liquid propane)
