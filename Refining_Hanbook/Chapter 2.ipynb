# Petroleum products and refinery configuration

***Description*** : A more in depth look at the petroleum products, their demand and environmental impacts.

## Petroleum products
### Fractions from the atomspheric and vacuum distillation
The cut length vary slightly depending on the crude oil source and finished product slate requirement.

#### Atomspheric overhead distillate
This consists of all the light materials of the crude absorbed into the overhead distillate. In most cases it forms the feed to the refineries light end units. Where LPGs, light naphta and heavy naphta are separated.

- ***Refinery Gas and LPGs*** : These are the C4s and lighter gases. They are either removed in the atmospheric tower or in the light end units depending on the refinery economic criteria. So first the de-butanizer, de-propanizer and finally the de-ethanizer.
- ***The naphtas*** : There are usually two naphta cuts : 
 - Light Naphtas (Light gasolines)
 - Heavy naphtas
Both of these products are the bottom product of the de-butanizer unit. They are separated in a naphta splitter fractionation tower.
- ***Straight run Kerosene*** : This fraction is usually the first side stream of a conventional atmospheric distillation column. It may be cut to meet a burning oil specification or to become a jet fuel component.
This kero fraction must meet a flash point specification after it is steam stripped and also a smoke point specification. Sulfur is dealt with by hydro-desulfurization, and the smoke point is reduced by removing or converting the aromatics. It can also be sent to the gas oil pool as a precursor for finished diesel product.
- ***Atmospheric straight run gas oils*** : Usually there will be two gas oil side streams, a light gasoil and below it a heavier one. Both are steam stripped to meet flash point specifications. The lighter stream is the precursor for automotive diesel grade finished product. The lower stream is usually a guard streal to correct diesel distillation end point. 
- ***Atmospheric residue*** : This is the bottom product of the atmospheric residue, the crude is not cut further to not risk cracking this fraction. The atmospheric towers that operate at higher cut points reduce the cracking by recycle of cold quench into the bottom of the tower and minimizing the residue hold up time.
The atmospheric residue may be routed to the fuel oil pool as the precursor of many grades of finisihed fuel oil products, the other options are : 
 - Feed to vacuum distillation
 - Feed to thermal cracker
 - Feed to a deep oil FCC
 - Feed to hydro-cracker or hydro-treater

#### Vacuum distillation of the atmospheric residue
This is acomplished under high vacuum conditions in specially designed towers that ensures a very lower pressure drops. usually there are two or three side streams : 
- ***HVGO*** : Which will be the feed to a hydro-cracker or FCC. A small heavier cut is taken off and returned to the bottom of the tower to correct the bottom distillate conradson carbon content.
- ***LVGO*** : This is sent to a hydro-desulfurizer and then either sent to gas oil pool as heating oil stock or routed to the fuel oil pool as blending stock
- ***The vacuum residue*** : It has many options to meet the refinety product slate.
 - In the case of energy refineries it can be upgraded to a prime distillate by thermal cracking or FCC or coking.
 - In the case of lube oils it can be processed to remove the heavy asphaltene portion of the stream. This de-asphaletened stream is called "Bright stock" and is excellent for lube oil blending stocks.
The asphaltenes has a wide spectrum of uses such bitumens used in buildings, road making.
