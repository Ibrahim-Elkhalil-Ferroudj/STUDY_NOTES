 #### ***What is machine learning?*** 
Machine learning is the science of programming computers to learn from data

to program a ML system ( ***Model*** ) you need to *Train it*. The examples ( *data* )  are called a *Training set*, and each example is called an *Instance*.

***Data mining*** is training a ML Algorithm using Large amounts of data. And then *inspecting* the algorithm to see the *patterns it has learned* ( _Basically helping human beings learn_ ).

#### ***What is machine learning great for?***
* Problems for which existing solutions require a lot of fine tuning or rules. You can simplify it using  *ML algorithms*
- Complex problems for which existing traditional approaches yields no good solutions 
- Fluctuating environments 
- Getting insights on complex problems and large amounts of data

##### ***Some of the algorithms and their usages in ML applications :***
- Image classification is typically performed using *C*onvolutional *N*eural *N*etworks (*CNN*)
- Natural language processing ( *NLP* ) or text classification which can be tackled using recurrent neural networks
- *NLP* Can be mixed with Natural Language Understanding ( *NLU* ) to create a chatbot or assistant. 
- Forecasting or predicting values is a regression task that can be tackled using many regression models. Such as *Linear* or *Polynomial* regression. regression *SVM* or *Random Forest* or an *Artificial Neural Network*. 
- There exists *Anomaly* and *Novelty* detection 
- To classify objects we use clustering 
- There is also data visualization which may be associated with dimensiolaity reduction techniques. 
- Building agents or bots is tackled using *Reinforcement Learning (RL)*. Where the agent is trained to pick actions ( *Policy* ) that will maximize their reward overtime.

#### ***Classification of Machine Learning systems***


##### Machine Learing systels are classified and categorized based on these criterias : 
1. Whether or not they are trained with human supervision (Supervised, Unsupervised, Semi-Supervised and RL).
2. Whether or not they can learn incrementally  on the fly (Online vs Batch learning).
3. Whether they work by simply comparing new data points to known data points, or instead by detecting patterns in the training data and building a predictive model (Instance vs model based learning).

In model based learning, to know which values will make your model performs best, you can either define a utility function ***To measure How good it is*** or a cost function ***to measure how bad it is***.

**It is crucial** to use a training set that is representative of the cases you want to ***Generalize to***.
*Ps :*
- Small trainings sets may result in sampling noise.
- A big training set but flawed sampling method may result in *Sampling Bias*.
 
*Most amount of effort data scientists do is cleaning the data.*
- If a  trained model doesn't generalize well it is possible *Overfitting*
A model overfits due to complexity of the model relative to hte amounts of noise in the training data. 
*Which Can be solve by :*
Simplifying the model by using one with fewer parameters, reducing the attributes or constraining the model ==> ***Regularization***
- Gather more Training data.
- Reduce the noise in the training data. 

There is also *Underfitting*, Opposite to the Overfitting. It occurs when the model is *too simple* to learn from the data. 
*The options you have here are :*
- Select a more powerful model with more parameters.
- Feed better features to the learning algorithm.
- Reduce the constraints on the model.
The amount of regularization to apply during learning can be controlled by a *Hyperparameter*.
A good method to see if the model works well is through ***Testing and Validation***. Where the data is split into two sets, *A training set* and a *Validation set*.
The error in the instances of the validation set is called *The generalization error*.
