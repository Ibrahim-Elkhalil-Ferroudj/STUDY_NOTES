We can say that a chemical reaction has taken place when a detectable number of molecules of one more species have lost their identity and assumed a new form.

To define a rate of reaction : Suppose A + 2B $ \rightarrow C + D$

the rate $-r_A > 0$ is the number moles of A consumed per unit of volume and per unit of time ($ \frac{mole}{dm^3  s} $)

In Hetergenous reaction systems, the rate of reaction is usually expressed in measures other than volume, such as reaction surface area or catalyst weight $ \frac{mole}{g(catalyst)  s} $

***The rate equation*** is an algebraic equation that is solely a function of the properties of the reacting materials and reaction conditions at a point in the system. It is independant of the reactor type.

***Example*** : $ -r_A = KC_A $ ; $ -r_A = KC^2_A $ ; $ -r_A = \frac{K_1C_A}{1+K_2C_A} $

For a given reaciton, the particular concentration dependance that the rate follows must be determined from experimental observations.

### Convection notation
- $ r_A $ is the rate of formation of the species *A*.
$ -r_A $ is the rate of disappearance.

## Mole Balances
### The general mole balance equation
To perform a mole balance first the system boundaries must be specified.
 A mole balance on species *j* on any time yields the following equation : 
 IN - OUT + Generation = Accumulation
 $F_{j0}$ - $F_{j}$ + $G_{j}$ = $\frac{dN_j}{dt}$

If all the system variables are spatially uniform throughout the sustem volume then $G_{j} = r_{ij}V$, otherwise in terms of subvolumes $G_{j} = r_{ij}\Delta V$

$\rightarrow G = \sum \Delta G = \sum r_{ij}\Delta V $ $ \rightarrow G = \int rdv $

The mole balance becomes : 
$F_{j0}$ - $F_{j}$ + $\int rdv$ = $\frac{dN_j}{dt}$

This is the general mole balance equation from which we can develop the design equations for the various types of industrial reactors. 

## Reactor types
### Batch reactor
This is used for small scale applications, ot for testing new processes that have not been fully developed yet or the manufacturing of expensive products. It has the advantage of high conversion but a drawback due the high labor cost. 

The mole balance for a batch reactor is as follows : $\frac{dN_j}{dt}$ = $\int rdv$

if perfect perfect agitation is assumed $\rightarrow \frac{dN_j}{dt} = rv $

### Continous flow reactors
These are almost always operated at steady state

#### 1. Continous Stirred Tank Reactor (CSTR)
Referred to as *Vat* or *Backmix reactor* and is used primarily for liquid phase reactions. It is assumed to be ideally mixed so when the mixing is not ideal we myst resort to ther modeling techniques such as residence time distribution. 

The mole balance for a CSTR is : $F_{j0}$ - $F_{j}$ + $\int rdv$ = 0
$\rightarrow V = \frac {F_{j0} - F_{j}}{-r_A} $ 

F = $C \cdot \vartheta$ $\rightarrow V = \frac {C_0 \cdot \vartheta_0 - C \cdot \vartheta}{-r_A} $

#### 2. Tubular reactors 
***Plug Flow Reactor (PFR)*** : 
These consist of a cylindrical pipe and operated at steady state. Most often used for gas phase reactions. For tubular reactors the reactants are continually consumed as the flow down the length of the reactor, so in modeling the reactor we assume the concentration varies continously in the axial direction throughout the reactor. Consequently, the rate of reaction which is a function of concentration for all but zero order reactions will vary axially.

For ***Plug Flow Reactors (PFR)*** : $F_{j0}$ - $F_{j}$ + $\int rdv$ = 0 $\rightarrow \frac{dF_j}{dV} = r_j $ 

$\rightarrow V = \int_{F_A}^{F_{A0} } \frac{dF_j}{-r_A}$

***Packed Bed reactor(PBR)*** :
 The principal difference between reactor design calcualtions invlolving single phase homogoneous and those involving two phase heterogenous reactions is that the latter takes place on the surface of the catalyst. Consequently, The reaction rate os based on the catalyst Mass ***W*** rather than the reactor volume.

For ***Plug Flow Reactors (PFR)***: 
$ W = \int_{F_A}^{F_{A0} } \frac{dF_j}{-r_A}$ (If pressure drop and catalyst decay are neglected)

## Industrial reactors
### 1. Liquid phase
Semi-Batch and CSTR are used primarily for liquid-phase reactions. A semi-batch reactor has the same disadvantages a Batch reactor has but it offers the advantage of temperatur control and minimizing unwanted reactions through feed control. The semi-Batch reactor is also used for two phase reactions in which gas is bubbled continously through the liquid. 
***A CSTR*** is used when intense agitation is needed where it can either be used on itself or in series or a battery of reactors.

### 2. Gas phase reactions 
The tubular reactor is relatively easy to maintain and offers the highest conversion per volume maong the continous reactors. However, it is difficult to maintain the temperature within the reactor and hot spots can occur when the reaction is exothermic. 

Hetergoneous gas phase reactions are usually carried out in PBRs. Same difficulties with temperature occurs, in addition the catalyst is troublesome to replace. On occasions channeling of gas flow occurs which results in inneffectiev use of reactor bed.

The analoguous reactor to CSTR that is catalytic is fluidized bed reactor where all it's contents are well mixed reslting in even temperature distribution.
